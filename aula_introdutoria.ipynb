Criar um script que irá auxiliar os alunos do SENAI a saberem se devem ou não comprar um uniforme.
Cursos acima de 40 horas, necessitam de uniforme. Até 40 horas, necessita crachá.
- Power BI: 32 horas;
- IA e Análise de Dados: 40 horas;
- Eletricista Instalador: 160 horas.

```python
power_bi = 32
ia = 40
eletricista = 160

#Cálculo para o curso de Power BI
if power_bi > 40:
    print("O curso de Power BI necessita de uniforme.")
else:
    print("O curso de Power BI necessita de crachá.")

#Cálculo para o curso de IA e Análise de Dados.
if ia > 40:
    print("O curso de IA e Análise de Dados necessita de uniforme.")
else:
    print("O curso de IA e Análise de Dados necessita de crachá.")
    
#Cálculo para o curso de Eletricista Instalador.
if eletricista > 40:
    print("O curso de Eletricista Instalador necessita de uniforme.")
else:
    print("O curso de Eletricista Instalador necessita de crachá.")
    
```

Verificar a qualidade das notas de um aluno.<br>
Maior do que 9: Excelente<br>
Maior do que 7: Boa<br>
Maior do que 5: Regular<br>
Abaixo de 5: Reprovado

```python
nota = int(input("Insira a sua nota: "))

if nota > 9:
    print("NOTA EXCELENTE!")
elif nota > 7:
    print("NOTA BOA")
elif nota > 5:
    print("NOTA REGULAR")
else:
    print("REPROVADO.")
```
