5. Montagem do gráfico

5.1 Montagem do gráfico de calor
- Para esse gráfico foi chamado a biblioteca ```seaborn``` (apelidada como ```sns```) e dele chamando a função `.heatmap()`, sendo responsável por gerar gráficos de calor.
- Dentro da função `.heatmap()` é preciso inserir algumas informações, sendo elas:
    - `data`: Aonde que esse gráfico vai puxar os dados, nesse caso, será puxado do `df[Colunas]` (sendo `Colunas` a lista com as colunas desejadas dessa base de dados), nessa base de dados será usado também a função `.corr()` para fazer uma correlação entre os dados dessa base.
    - `anott`: Quando estiver como `True`, cada célula do mapa de calor terá um valor indicando se o valor foi muito terminante pro fator target ou não.
        - Números entre -1 e 0, quer dizer que a relação é inversamente proporcional. (Nesse caso, se estive em um número negativo, o fator contribuiu negativamente pra nota do exame).
        - Se a célula estiver mais próximo de 0 ou o próprio 0, aquele fator não contribui em nada para o target desejado.
        - Para números de 0 à 1, quer dizer o quanto aquele fator contribui positivamente para o target desejado (ou seja, nesse exemplo, quanto mais próximo do 1, mais contribuiu para o aluno receber uma nota melhor).
    - `fmt`: Como será a formatação da célula, nesse caso foi usado `.2f`, ou seja, será exibido somente 2 números de casa decimal.
    - `cmap`: O mapa de cores usado para gerar o heatmap, nesse caso foi usado o `coolwarm`.

````python
#Gerar gráfico heatmap
sns.heatmap(data = df[Colunas].corr(), annot = True, fmt = ".2f", cmap = "coolwarm");
````

5.2.1 Análise das horas de estudo (fator considerado importante para contribuir com o resultado do exame)
- Através das horas de estudo foram feitas as médias entre os alunos que estudaram de 7 horas pra mais e os que estudaram menos de 4 horas.

````python
#Filtrar quantidade de horas de estudo.

filtro_horas_altas = df["study_hours_per_day"] >= 7

filtro_horas_baixas = df["study_hours_per_day"] < 4
````

- Para obter a média foi feito o seguinte:
    - Primeiro foi feito a filtragem de quais estudantes estudaram de 7h pra cima e quais estudaram menos de 4 horas.
        - Para os que estudaram mais de 7h, esses dados foram separados na variável `filtro_horas_altas`.
        - Para os que estudaram menos de 4h, esses dados foram separados na variável `filtro_horas_baixas`.
    - Após isso foi feito a média com a nota dos estudantes de cada uma dessas categorias usando a função `.mean()`
        - É chamado o dataframe `df`, pedindo que, dessa base de dados, sejam filtrados somente os alunos do filtro específico (`filtro_horas_altas` ou `filtro_horas_baixas`), da base de dados será usado a coluna `exam_score` e com isso será feito a média com o `.mean()`.
        - Portando, o comando será `df[filtro]["exam_score"].mean()`
    - A média de cada um desses filtros foi atribuído em uma variável para facilitar na hora de escrever uma mensagem personalizada para o usuário, relatando o resultado da filtragem.

````python
media_pessoas_estudam_muito = df[filtro_horas_altas]["exam_score"].mean()
media_pessoas_estudam_pouco = df[filtro_horas_baixas]["exam_score"].mean()

print(media_pessoas_estudam_muito)
print(media_pessoas_estudam_pouco)

print(f"\nA pontuação média das pessoas que estudam mais de 7h por dia é de {media_pessoas_estudam_muito:.2f}, as que estudaram menos de 4h, a pontuação média é de {media_pessoas_estudam_pouco:.2f}, logo, para obter um resultado melhor é preciso estudar mais de 8h por dia.")
````

A pontuação média das pessoas que estudam mais de 7h por dia é de 99.38, as que estudaram menos de 4h, a pontuação média é de 60.90, logo, para obter um resultado melhor é preciso estudar mais de 8h por dia.

5.3 Criação do gráfico de regressão + dispersão
- Para a criação desse gráfico, será usado a função `.lmplot()`, gerando um gráfico de regressão junto com a dispersão dos dados (onde cada dado será ilustrado como uma bolinha, sendo alocado dentro do gráfico de acordo com a relação).

---
5.3.1 Análise da relação entre as horas de estudo por dia e a nota do exame
- Nessa análise é possível ver a proporção que, quanto mais horas estudadas por dia, maior é a tendência de obter um bom desempenho no exame
- É possível ver também várias bolinhas nesse gráfico, sendo elas, como foi a colocação de cada estudante em relação as horas de estudo por dia.
