```python
#Fazer o tratamento da base de dados
df = pd.read_csv(r"data\dataset_aula01_techshop_vendas.csv", parse_dates=["data_pedido"])
#Exibir resultado para testar se deu certo.
df.info()
```

```python
#Criar uma cópia da database que já foi tratada ou manipulada.
df_tratado = df.copy()
```

```python
#Adição da coluna "faturamento" na database já tratada.
df_tratado["faturamento"] = df_tratado["preco_unitario"] * df_tratado["quantidade"]
df_tratado.head()
```

```python
#Fazer a soma de todos os valores presentes em uma coluna, nesse caso, a soma da coluna faturamento.
df_tratado["faturamento"].sum()
```

```python
#É possível também fazer um print dessa operação, assim eliminando algumas informações desnecesárias.
print(df_tratado["faturamento"].sum())
```

```python
#É possível também atribuir essa operação em uma variável
faturamento_total = df_tratado["faturamento"].sum()
print(f"O faturamento é de R${faturamento_total}.")
```

```python
#Formatação da variável, executando replaces para as substituições adequadas.
faturamento_correto = f"{faturamento_total:,.2f}".replace(",","x").replace(".",",").replace("x",".")
print(f"O faturamento total foi de R${faturamento_correto}.")
```

```python
#Realizar agrupamentos para realizar os faturamentos por categoria
df_tratado.groupby("categoria")["faturamento"].sum()
```

```python
#Realizar o agrupamento e também organizar em ordem crescente.
df_tratado.groupby("categoria")["faturamento"].sum().sort_values()
```

```python
#Agrupamento com base por região.
df_tratado.groupby("regiao")["faturamento"].sum().sort_values()
```

```python
#Agrupamento com base por região em ordem descrescente.
df_tratado.groupby("regiao")["faturamento"].sum().sort_values(ascending=False)
```

```python
df_tratado.groupby(["regiao","categoria"])["faturamento"].sum()
```

```python
#Criar função para formatar os valores
def converter(valor):
    return f"R${valor:,.2f}".replace(",","x").replace(".",",").replace("x",".")
```

```python
#Teste da função de conversão.
agrupamento_faturamento_regiao = df_tratado.groupby("regiao")["faturamento"].sum()
agrupamento_faturamento_regiao.map(converter)
```

```python
#Move o dataset modificado para o dataframe padrão.
df = df_tratado.copy()
```

```python
#Gera o arquivo tratado dentro da pasta.
df.to_csv("dataset_tratado.csv")
```
