---
Teste de previsão usando o livro `Dom Casmurro`.

---

Fazendo a importação do livro que está em um arquivo `.txt` e a tratativa para que todas as palavras estejam em minúsculo com o `.lower()`

```python
arquivo = open(r"data\dom_casmurro.txt", encoding = "utf-8")
texto = arquivo.read().lower()
```

```python
palavras = texto.split()
palavras
```

Convertendo todo o livro para um dataframe e fazendo a exibição das 5 primeiras linhas usando o comando `.head()`.

```python
import pandas as pd
tabela = pd.DataFrame({"palavra":palavras})
tabela.head()
```

fazendo o sistema de `próxima palavra` para auxiliar na análise de qual é a próxima.

```python
tabela["proxima_palavra"] = tabela["palavra"].shift(-1)
tabela
```

Tratativa para a remoção da última linha que estava com um `NaN`.

```python
tabela = tabela.dropna()
tabela
```

---
Filtrando uma palavra qualquer e exibindo qual é a palavra que vem depois dela dentro do texto.

```python
filtro = tabela["palavra"] == "que"
tabela[filtro]
```

A exibição em porcentagem de quantas vezes a `proxima_palavra` aparece depois da palavra filtrada.

```python
tabela[filtro]["proxima_palavra"].value_counts(normalize = True)
```

Exibição da palavra que mais tem chance de aparecer depois da palavra filtrada.

```python
tabela[filtro]["proxima_palavra"].value_counts().idxmax()
```
