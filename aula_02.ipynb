<h1>Estruturas de Dados</h1>

A Lista permite operações

```python
#Lista
temperaturas = [30,35,39,40,42,45,38,37]
print(temperaturas)
```

```python
#Filtragem da lista com base na posição de algum item
print(temperaturas[0])
print(temperaturas[3])
```

```python
print(f"O primeiro registro foi de {temperaturas[0]}°C!")
```

```python
#Mudança de valores de algum item da lista
temperaturas[0] = 100
print(temperaturas)
```

```python
#Adição de itens na lista
temperaturas.append(67)
print(temperaturas)
```

```python
#Remoção de itens na lista
temperaturas.remove(39)
print(temperaturas)
```

```python
#Remoção de itens com base na posição do item
temperaturas.remove(temperaturas[0])
print(temperaturas)
```

```python
#Organização dos itens da lista para ordem decrescente.
temperaturas.sort(key = int, reverse = True)
print(temperaturas)
```

<h3>Desafio</h3>
Criar uma lista de compras que solicite que o usuário insira até 3 itens para uma lista de compras. Exibir a lista de compras.

```python
lista_compras = []
indice_lista = 0

while True:
    indice_lista += 1
    
    item_lista = input(f"Digite o {indice_lista}º item da lista de compras: ")
    # item_lista = input("Digite o nome do item desejado para a lista de compras: ")
    lista_compras.append(item_lista)
    
    print(f"Segue a lista atual: {lista_compras}")
    confirmacao = input("Deseja continuar? ")
    if confirmacao.lower() in ['n', 'nao', 'ñ', 'não']:
        print(f"Segue a lista de compras finalizada: {lista_compras}")
        break
```
