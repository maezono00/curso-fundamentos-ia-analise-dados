Conceito de n-gramas

```python
frase = "O amor é cego o amor é forte e o amor vence tudo"
```

Função `.split()` é usada para pegar todas as palavas de uma string e agrupar a frase numa lista.

```python
palavras = frase.split()
palavras
```

Importação da biblioteca `pandas` para montar um dataframe usando a lista de palavras geradas.

```python
import pandas as pd
tabela = pd.DataFrame({"palavra":palavras})
tabela
```

Uma representação gráfica de previsão de palavras.

```python
tabela["proxima_palavra"] = tabela["palavra"].shift(-1)
tabela
```

Tratativa para remover a última linha que estava com um `NaN`.

```python
tabela = tabela.dropna()
tabela
```

Filtragem para exibir os casos que tem a palavra `amor` e as palavras que vem depois dele.

```python
filtro = tabela["palavra"] == "amor"
tabela[filtro]
```

Contagem das palavras que vem depois de `amor`, fazendo a exibição de porcentagem das respectivas palavras.

```python
tabela[filtro]["proxima_palavra"].value_counts(normalize = True)
```

Função para chamar qual palavra que mais tem chance de aparecer depois do `amor`.

```python
tabela[filtro]["proxima_palavra"].value_counts().idxmax()
```

Testes para ver qual será a próxima palavra

```python
filtro = tabela["palavra"] == "amor"
tabela[filtro]
```

```python
tabela[filtro]["proxima_palavra"].value_counts(normalize = True)
```

```python
tabela[filtro]["proxima_palavra"].value_counts().idxmax()
```
