Criar duas listas:<br>
- 5 produtos (apenas inserir valores)<br>
- A primeira lista, sendo o custo do produto.<br>
- A segunda lista, o preço de venda<br>

O desafio é calcular o lucro de cada produto.

```python
preco_custo = [10,20,30,40,50]
preco_venda = [67,87,107,127,147]

array_custo = np.array(preco_custo)
array_venda = np.array(preco_venda)

#Usando subtract
# array_lucro = np.subtract(array_venda, array_custo)

#Fazendo subtração
array_lucro = array_venda - array_custo

indice = 0
for lucro in array_lucro:
    indice += 1
    print(f"O lucro do produto {indice} foi de: R${lucro}")

# print(array_lucro)
# print(array_lucro[0])
```

```python
salarios = np.array([2,3,4,5,6,500])
maior_salario = np.max(salarios)

print(maior_salario)
```

```python
menor_salario = np.min(salarios)

print(menor_salario)
```

```python
media_salarios = np.mean(salarios)
print(f"{media_salarios:.2f}")
```

```python
mediana_salarios = np.median(salarios)
print(mediana_salarios)
```

```python
desvio = np.std(salarios)
print(desvio)
```

Criar uma lista de vendas por dia<br>
vendas = [10,20,45,5,90]<br>
- Exibir a maior venda<br>
- Exibir a menor venda<br>
- Exibir a média de vendas<br>
- Exibir a mediana de vendas

```python
array_vendas = np.array([10,20,45,5,90])
```

```python
maior_venda = np.max(array_vendas)
print(f"A maior venda foi: {maior_venda}")
```

```python
menor_venda = np.min(array_vendas)
print(f"A menor venda foi: {menor_venda}")
```

```python
media_venda = np.mean(array_vendas)
print(f"A média das vendas foi de: {media_venda}")
```

```python
mediana_venda = np.median(array_vendas)
print(f"A mediana das vendas foi de: {mediana_venda}")
```
